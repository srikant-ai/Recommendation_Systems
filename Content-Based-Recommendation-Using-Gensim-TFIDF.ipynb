```python
import pandas as pd
import numpy as np

movieDf= pd.read_csv("D:\Recommendation_Systems\movies_keywords.csv")
```

```python
# Content Based Filtering
```

```python
movieDf["keywords"]=movieDf["keywords"].apply(lambda x : x.lower().split(","))
```

```python
# Need to list of keywords in each movie 
keyword_list=[]
for record in movieDf["keywords"]:
    keyword_list.append(record)
```

```python
keyword_list
```

```python
pos=1
for element in keyword_list:
    print("Movie {} has {}".format(pos,len(element)))
    pos=pos+1
```

```python
from gensim.corpora.dictionary import Dictionary
"""
Dictionary encapsulates the mapping between normalized words and their integer ids.
"""
dictionary = Dictionary(keyword_list)
print(dictionary)
```

```python
dictionary[0] , dictionary[1] , dictionary[2]
```

```python
"""
The main function is doc2bow, which converts a collection of words to its bag-of-words representation: 
a list of (word_id, word_frequency) 2-tuples.
"""
bow = dictionary.doc2bow(keyword_list[0])
bow
```

```python
"""
Using List Comprehension to create a Corpus of bag of Words for Each Document i.e keywords of Each Individual Movie.
"""
corpus = [dictionary.doc2bow(doc) for doc in keyword_list]
corpus
```

```python
"""
Objects of this class realize the transformation between word-document co-occurrence matrix (integers) 
into a locally/globally weighted TF_IDF matrix (positive floats).
VocabTransform(old2new[, id2token])	Remap feature ids to new values.

Source : https://radimrehurek.com/gensim/models/tfidfmodel.html
"""
from gensim.models.tfidfmodel import TfidfModel
tfidf = TfidfModel(corpus)
tfidf
```

```python
print(tfidf)
```

```python
print(tfidf[corpus])
```

```python
from gensim.similarities import MatrixSimilarity
"""
Compute similarity against a corpus of documents by storing the index matrix in memory. 
The similarity measure used is cosine between two vectors.

Use this if your input corpus contains dense vectors (such as documents in LSI space) and fits into RAM.

The matrix is internally stored as a dense numpy array. Unless the entire matrix fits into main memory, 
use Similarity instead.
"""

sims = MatrixSimilarity(tfidf[corpus], num_features=len(dictionary))
print(sims)
```

```python
similarity_array=sims[corpus[0]]
```

```python
pd.Series(similarity_array.tolist(), index=movieDf.primaryTitle.values).sort_values(ascending=False)[:]
```
